```python
df.rename(columns={
    "total_cases": "Confirmed",
    "total_deaths": "Deaths",
    "total_recovered": "Recovered"
}, inplace=True)
```

```python
def safe_rename(dframe):
    mapping = {}
    for col in dframe.columns:
        cl = col.lower()
        if cl in ["country", "country/region", "country_region", "location"]:
            mapping[col] = "Country"
        elif cl in ["confirmed", "cases", "total_cases"]:
            mapping[col] = "Confirmed"
        elif cl in ["deaths", "total_deaths"]:
            mapping[col] = "Deaths"
        elif cl in ["recovered", "total_recovered"]:
            mapping[col] = "Recovered"
        elif cl == "date":
            mapping[col] = "Date"
    dframe.rename(columns=mapping, inplace=True)

safe_rename(df)
```

```python
print(df.columns)
```

```python
n_countries = df["Country"].nunique()
print("Unique countries:", n_countries)
```

```python
top_confirmed = (
    df.groupby("Country")["Confirmed"]
      .max()
      .sort_values(ascending=False)
      .head(10)
)
top_confirmed
```

```python
df["Date"] = pd.to_datetime(df["Date"])

global_daily = df.groupby("Date")["Confirmed"].sum()

plt.figure(figsize=(12, 6))
plt.plot(global_daily, color="blue")
plt.title("Global Confirmed COVID19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_deaths = (
    df.groupby("Country")["Deaths"]
      .max()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_deaths.values, y=top_deaths.index, palette="Reds_r")
plt.title("Top 10 Countries by Total COVID19 Deaths")
plt.xlabel("Total Deaths")
plt.ylabel("Country")
plt.show()
```

```python
total_recovered = df["Recovered"].sum()
total_deaths    = df["Deaths"].sum()

plt.figure(figsize=(6, 6))
plt.pie(
    [total_recovered, total_deaths],
    labels=["Recovered", "Deaths"],
    autopct="%1.1f%%",
    colors=["green", "red"],
    startangle=90,
)
plt.title("Global COVID19 Outcomes")
plt.show()
```
